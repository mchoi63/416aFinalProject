### 2.1 Create the Network (10%)
(i) First partially build this vast network. Select only the edges that refer to the “beneficiary-of” relationship (there are 19,194 edges labeled `Beneficiary of` and `beneficiary of`). 

Remember that when you add an edge to a network, NetworkX also adds both incident nodes. However, at the moment, the nodes have only more or less randomly chosen labels—and no attributes. Let’s import the attributes and true names from the other three files.

Then, you need to go through the node files for entities, officers, and intermediaries and search for the node information. For each node, you will store its name, type, and a three-letter country code. The first code block should import all the necessary modules and define the constants.

The purpose of the node dictionary is to facilitate future lookup. (Python lists have linear lookup time.) Read each of the files with a CSV dictionary reader and extract and collect the desired attributes. Note that there is no need to process rows that do not match any existing node (because your network does not include all nodes and edges) and add any nodes to the graph (because they have been already added by way of the incident edges). When done, update the node attributes "country" and "kind", and relabel the nodes to match persons and organizations names.


**Your Program Below:** 

```python
##### Our code is based off of CNA Chapter 9, as suggested above in the notes of this proble #####
# Import necessary packages
import csv
import numpy as np
import pandas as pd
import networkx as nx
import itertools
import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd
import math
import sys
reload(sys)
sys.setdefaultencoding('utf8')

# Make an empty graph
gPanama = nx.Graph()

# Open the csv file

with open("data/csv_panama_papers/panama_papers.edges.csv") as file:
    # Read it with DictReader, as provided in the instruction
    dataDict = csv.DictReader(file)
    # Iterate through each line
    for line in dataDict:
        # If the Link starts with "Beneficiary of" / "beneficiary of"
        if line["link"].lower().startswith("beneficiary of"):
            # Add the edge
            gPanama.add_edge(line["START_ID"], line["END_ID"])   #I GOT RID OF THIS!

# Check if the number of nodes are correct
# (this is different from the instructions above, but this is the same with the confirmed Piazza post)
print("Number of edges: ")
print(gPanama.number_of_edges())

# Get nodes without duplicates
graphNodes = set(gPanama.nodes())

# Get the node data
for filename, countryCode, name in (("data/csv_panama_papers/panama_papers.nodes.entity.csv", "jurisdiction", "name"),
                               ("data/csv_panama_papers/panama_papers.nodes.officer.csv", "country_codes", "name"),
                               ("data/csv_panama_papers/panama_papers.nodes.intermediary.csv", "country_codes", "name")):
    with open(filename,'U') as infile:
        # Get the kind of the data (based on filename)
        kind = filename.split(".")[2]
        # Get the data from each file, using DictReader
        data = csv.DictReader(infile)
        # Dictionary for country names
        country_names = {}
        # Iterate through the data to get the country names of the nodes in the graph
        for line in data:
            if line["node_id"] in graphNodes:
                country_names[line["node_id"].strip()] = (line[name].strip().upper(), line[countryCode])
    
    # Dictionaries for names, countries, and kinds
    names = {}
    countries = {}
    kinds = {}
    
    # Iterate through country names to fill out the dictionaries
    for node_id, value in country_names.items():
        names[node_id] = value[0]
        countries[node_id] = value[1]
    for node_id, _ in country_names.items():
        kinds[node_id] = kind
    
    # Set attributes of kind, name and country for each nodes
    nx.set_node_attributes(gPanama, kinds, "kind")
    nx.set_node_attributes(gPanama, names, "name")
    nx.set_node_attributes(gPanama, countries, "country")
    
#Get rid of unnecessary nodes
if "ISSUES OF:" in gPanama:
    gPanama.remove_node("ISSUE OF:")
if "" in gPanama:
    gPanama.remove_node("")

print("nodes:")
print(gPanama.number_of_nodes())

print("edges:")
print(gPanama.number_of_edges())
```

### 2.3 Analyze the Network: Degree Distribution (10%)
(i) Compute the degree distribution and plot it using log-log scale.

**Your Program Below:** 

```python
# Compute degree of each node
degree = dict(nx.degree(gPanama))

# Make another dictionary based on the distribution of degrees
degree_dist = {}
for v in degree.values():
    if v in degree_dist.keys():
        degree_dist[v] = degree_dist[v] + 1
    else:
        degree_dist[v] = 1

# Create a log log plot
plot = plt.figure()
scatterplot = plt.gca()

scatterplot.scatter(degree_dist.keys(), degree_dist.values())
plt.xlabel("degree")
plt.ylabel("frequency")
scatterplot.set_yscale('log')
scatterplot.set_xscale('log')
```

### 2.4 Analyze the Network: Attributes and Assortativity (15%)
(i) What are the node attributes in the data and how many classes (categories) do they each have? 

**Your Program Below:** 

```python
# Get the connected components with more than 20 edges or nodes
components = []

for subgraph in nx.connected_component_subgraphs(gPanama):
    if nx.number_of_edges(subgraph) >= 20 or nx.number_of_nodes(subgraph) >= 20:
        components.append(subgraph)

# Create a subgraph based on the components
sub_panama = gPanama.subgraph(itertools.chain.from_iterable(components))

# Make a dictionary of attributes and classes
attr_dict = {}

# Iterate through, and for each "attribute" put in classes
for node in sub_panama.nodes.data():
    for i in node[1]:
        if i in attr_dict.keys():
            if node[1][i] not in attr_dict[i]:
                attr_dict[i].append(node[1][i])
        else:
            attr_dict[i] = list()
            attr_dict[i].append(node[1][i])

# Make a dictionary with attributes and counts of classes
attr_count = {}

##### We were just curious what this would look like (or we can reproduce something similar to the graph on the book)
##### So we tested it here! (and we have it left up here since it was cool!)

# Iterate through the dictionary of attributes and classes, and count the number of classes per attribute
for k in attr_dict.keys():
    attr_count[k] = len(attr_dict[k])
    
    
color_dict = {"entity":"yellow", "officer":"black", "intermediary":"green"}

# Assign color for each node based on the color dictionary
nodeColors = []


nodes = sub_panama.nodes()

colors = [color_dict[sub_panama.nodes[n]['kind']] for n in nodes]

#drawing nodes and edges separately so we can capture collection for colobar
pos = nx.spring_layout(sub_panama)



ec = nx.draw_networkx_edges(sub_panama, pos, alpha=0.8, width=0.5)

nc = nx.draw_networkx_nodes(sub_panama, pos, nodelist=nodes, node_color=colors, 
                            with_labels=False, node_size=0.2, cmap=plt.cm.jet)



plt.savefig("miniData.pdf")


attr_count
```

Based on your findings above, is the network assortative with respect to the attribute "kind"? 

**Answer:** 

No, the network is perfectly disassortative with respect to the node kinds.

(iii) How does NetworkX compute the attribute assortativity coefficient? Describe all variables in your formula. (You can type formulas in latex using the \$ environment.) 

**Answer:** 

The coefficient that the NetworkX computes for the attribute assortativity coefficient is $r = \frac{\sum_{xy}xy(e_{xy}-a_{x}b_{y})}{\sigma_{a}\sigma_{b}}$, where $e_{xy}$ is the fraction of all edges in the network that join together nodes with attributes x and y, and $a_{x}$ is the fraction of edges that connect the nodes with the attribute x and $b_{y}$ is the fraction of edges that connect the nodes with the attribute y.

 Compute the assortativity coefficient of the attribute "kind" and the degree assortativity coefficient.

**Your Program Below:** 

```python
asso_coeff = nx.attribute_assortativity_coefficient(sub_panama, "kind")
deg_coeff = nx.degree_assortativity_coefficient(sub_panama)

print("Assortativity coefficient:")
print(asso_coeff)
print("Degree assortativity coefficient:")
print(deg_coeff)
```

Compare the assortativity coefficient of the attribute "kind" with the degree assortativity coefficient. What are your conclusions about the money-laundering connections of entities and officers? 

**Answer:**

The attribute assortativity coefficient is -1, which means that the network is completely dissortative with respect to the attribute "kind". On the other hand, the degree assortativity coefficient is a negative value, not as small as -1. Since it is negative, we know that the nodes with higher degree are, on average, surrounded by the nodes with lower degree, and vice versa. 

Since the network is completely disassortative with respect to the attribute "kind", this means that the officers are connected to the entities, but not to other officers, and similar situation holds to the entities: they are connected to the officers, not to other entities. This means that the off-shore money-laundering happens between the officers and entities, but not in between officers or entities. Also, since the degree assortativity coefficient is negative, this means that there exist central players (the hubs) in the money-laudering connections.
